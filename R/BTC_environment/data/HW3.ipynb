```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("export.csv")
```

```python
df.head()
```

```python
df2 = df.groupby("country").agg(np.mean)
df2 = df2.rename(columns={"monthly_absolute_hashrate_EH/S":"monthly_avg_hashrate"},index={"Germany *":"Germany", "Ireland *":"Ireland"})
df2
```

```python
df2.sort_values("monthly_avg_hashrate", inplace=True)
ax = df2.plot.barh(figsize=(20,11),title="Average monthly hashrate EH/s by country",xticks=(range(0,70,5)))
ax.set_xlabel("EH/s")
ax.set_ylabel("countries")
ax.grid(axis="x",linestyle="--")
```

```python
ax.get_figure().savefig("DobryWykres.jpg",dpi=80)
```
